### Rotten Tomatoes 

#### I will start by having a look at the data from RT, and comparing the user rating with their own critics rating. This can eb done with a scatterplot.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=all_sites, x='RottenTomatoes', y='RottenTomatoes_User');
```

#### I will now quantify this difference by subtracting the user rating from the critic's rating. The more positive the difference the more the critics liked the film compared to the user and visa versa. I will make a new column with this difference.

```python
all_sites['Rott_diff'] = all_sites['RottenTomatoes'] - all_sites['RottenTomatoes_User']
all_sites
```

#### I would like to know the average amount the critics and the users disagreed, for this I will need to take the absolute value of the difference and then take the mean.

```python
all_sites['Rott_diff'].abs().mean()
```

#### Below I will plot the distribution of the differences between the RT critics score and the RT user score.

```python
plt.figure(figsize=(12,4))
sns.histplot(data=all_sites, x='Rott_diff', bins=25, kde=True, color='magenta')
plt.title('RT Critics Score Minus RT User Score');
```

#### Now I will do the same but for the absolute value.

```python
plt.figure(figsize=(12,4))

sns.histplot(data=np.abs(all_sites['Rott_diff']), bins=25, kde=True, color='magenta')
plt.title('Absolute Value of RT Critics Score Minus RT User Score');
```

### Fandango vs All Sites

#### Now we will start exploring both data sets and comparing them.

#### First i will merge both data sets based on the FILM column via an inner merge, this should give us back a dataframe with where only the films that are in both data sets exist. I will need to use the original Fandango dataframe before I stripped the year.

```python
fandango = pd.read_csv('fandango_scrape.csv')
```

```python
merged_data = pd.merge(fandango, all_sites, on='FILM', how='inner')
merged_data.head()
```

#### As it stands, it isn't as simple as it should be to compare ratings across each column, i will need to normalise the columns that don't have a 0-5 rating system.

```python
merged_data['RottenTomatoes'] = merged_data['RottenTomatoes'] * (5 / 100)
merged_data['RottenTomatoes_User'] = merged_data['RottenTomatoes_User'] * (5 / 100)
merged_data['Metacritic'] = merged_data['Metacritic'] * (5 / 100)
merged_data['Metacritic_User'] = merged_data['Metacritic_User'] * (5 / 10)
merged_data['IMDB'] = merged_data['IMDB'] * (5 / 10)

merged_data.head()
```

#### Let's make a dataframe with only the ratings as columns.

```python
ratings = merged_data[['STARS', 'RATING', 'RottenTomatoes', 'RottenTomatoes_User',
                       'Metacritic', 'Metacritic_User', 'IMDB']]
ratings
```

### Now it is time to see if Fandango ratings are higher than what they probably should be.

#### I will create a KDE plot with all the ratings on the same graph so we can see how the ratings are distributed.

```python
plt.figure(figsize=(14,6))

for column in ratings.columns:
    sns.kdeplot(ratings[column], label=column, clip=(0,5), fill=True)

plt.legend(loc='upper left');
```

#### it seems that Fandango has an uneven distribution but also Rotton Tomatoes critics has the most even distribution. Let's compare the two.

```python
plt.figure(figsize=(10,5))

sns.kdeplot(data=ratings['RottenTomatoes'], label='RottenTomatoes', clip=(0, 5), fill=True)
sns.kdeplot(data=ratings['STARS'], label='STARS', clip=(0, 5), fill=True)

plt.legend(loc='upper left');
```

#### Below is a histagram of all the ratings.

```python
plt.figure(figsize=(14,6))
sns.histplot(data=ratings, bins=50);
```

#### With a cluster map, it should cluster together all the poorly rated films and all the highly rated films, but as we can see below, for the Fandago ratings there is a clear colour difference, tellin us that Fandango ratings do not agree with the majorty of the other ratings.

```python
sns.clustermap(ratings);
```

#### I would like to see what are the top ten worst rated films rated by Rotten Tomatoes critics, for this i will add back the FILMS column and then show the dataframe in ascending order in terms of the RT column.

```python
# add back FILM column and move it to first column index
ratings['FILM'] = merged_data['FILM']
ratings.insert(0, 'FILM', ratings.pop('FILM'))

top_ten_worst = ratings.sort_values(by='RottenTomatoes')[:10]
top_ten_worst
```

#### As we can see with the kde plot below, once again the Fandago ratings and stars appear much higher even for poorly rated films.

```python
plt.figure(figsize=(12,6))

for column in top_ten_worst.drop('FILM', axis=1).columns:
    sns.kdeplot(top_ten_worst[column], label=column, clip=(0,5), fill=True)

plt.legend();
```

#### Let's find the film in which they have bumped the score up the most, for this i will subtract the RT critics column from the Fandango ratings column then find the max value.

```python
pd.DataFrame(ratings.loc[(ratings['RATING'] - ratings['RottenTomatoes']).idxmax()]).T
```

#### A bar plot will help us visualise the difference.

```python
plt.figure(figsize=(14,7))
plt.title('Taken 3 Ratings Out of 5')
sns.barplot(pd.DataFrame(ratings.loc[(ratings['RATING'] - ratings['RottenTomatoes']).idxmax()]).T, palette='colorblind');
```
